```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Sales_data_zone_wise.csv')
```

# ***THE MEAN SALE GENERATED BY EACH ZONE***

```python
zoneA=data['Zone - A'].array
print(zoneA)
```

```python
zoneB=data['Zone - B'].array
zoneC=data['Zone - C'].array 
zoneD=data['Zone - D'].array
```

```python
print('meanof zoneA =',zoneA.mean())
print("mean of zoneB =",zoneB.mean())
print('mean of zoneC =',zoneC.mean())
print('mean of zoneD =',zoneD.mean())
```
