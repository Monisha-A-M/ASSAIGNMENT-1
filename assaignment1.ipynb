```python
plt.figure(figsize=(8,6))

plt.subplot(2,1,1)
plt.scatter(data['PL'],data['PW'],s=10,marker='*',c='red')
plt.title("scatter plot of PL Vs PW")
plt.xlabel("PL")
plt.ylabel("PW")

plt.subplot(2,1,2)
plt.scatter(data['SL'],data['SW'],s=10,marker='*',c='green')
plt.title("scatter plot of SL Vs SW ")
plt.xlabel("SL")
plt.ylabel("SW")
plt.tight_layout()
```
